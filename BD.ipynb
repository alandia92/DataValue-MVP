```python
import random
from datetime import datetime, timedelta

# Conexión al servidor de MongoDB (asegúrate de tener MongoDB en ejecución)
myclient = pymongo.MongoClient("mongodb+srv://arratelandia92:Y7QvVogQcF2iMWPq@cluster0.tx3uwk0.mongodb.net/test")

# Acceso a la base de datos
db = myclient["database"]

# Acceso a la colección "ayudas"
ayudas_collection = db['ayudas']

# Crear una instancia de Faker para generar datos aleatorios
fake = Faker('es_ES')

# Categorías de ayuda
categorias = ['tramites', 'hogar', 'ayuda movilidad']

# Subcategorías de la categoría "tramites"
subcategorias_tramites = ['tecnologia', 'tramites y ayudas', 'vamos al banco', 'tramites medicos']

# Subcategorías de la categoría "hogar"
subcategorias_hogar = ['compras', 'pequeñas reparaciones', 'nutricion y alimentacion']

# Subcategorías de la categoría "ayuda movilidad"
subcategorias_movilidad = ['transporte', 'ocio']

# Rango de fechas para el año 2022
fecha_inicio = datetime(2022, 1, 1)
fecha_fin = datetime(2022, 12, 31)

# Generar datos aleatorios de ayudas
for _ in range(50000):
    usuario_id = random.randint(0, 10000)
    fecha = fake.date_between_dates(date_start=fecha_inicio, date_end=fecha_fin)
    fecha_str = fecha.strftime("%Y-%m-%d")  # Convertir la fecha a una cadena de texto en formato 'YYYY-MM-DD'
    categoria = random.choice(categorias)

    # Generar subcategorías según la categoría
    if categoria == 'tramites':
        subcategoria = random.choice(subcategorias_tramites)
    elif categoria == 'hogar':
        subcategoria = random.choice(subcategorias_hogar)
    elif categoria == 'ayuda movilidad':
        subcategoria = random.choice(subcategorias_movilidad)

    ayuda = {
        'usuario_id': usuario_id,
        'fecha': fecha_str,
        'categoria': categoria,
        'subcategoria': subcategoria
    }

    ayudas_collection.insert_one(ayuda)


# Cerrar la conexión
myclient.close()
```
